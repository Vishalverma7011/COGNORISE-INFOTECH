```python
import seaborn as sns
import matplotlib.pyplot as plt

def bar_plot(feature):
    plt.figure(figsize=(9, 4))
    
    sns.countplot(x=feature, data=df, palette="viridis")
    
    plt.ylabel("Frequency")
    plt.title(feature)
    plt.show()
    
    varV = df[feature].value_counts()
    print(f"{feature}:\n{varV}")

categorical_variables = ["Marital_Status", "Complain"]

for variable in categorical_variables:
    bar_plot(variable)
```

```python
plt.figure(figsize=(9, 4))
sns.histplot(df["Income"], bins=100, kde=True, color='skyblue')
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.title("Income Distribution")
plt.show()
```

```python
def plot_hist(feature):
    plt.figure(figsize=(9, 4))
    sns.histplot(df[feature], bins=50, kde=True, color='skyblue')
    plt.xlabel(feature)
    plt.ylabel("Frequency")
    plt.title(f"Distribution of {feature}")
    plt.show()

numeric_variables = ["Year_Birth", "Kidhome", "Teenhome"]

for variable in numeric_variables:
    plot_hist(variable)
```
